```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

from flask import Flask, jsonify
import datetime as dt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. '%Y-%m-%d'
#p = '20\d{2}-06-\d{2}'
m = "06"
results = session.query(Measurement.date, Measurement.tobs).\
    filter(func.strftime("%m", Measurement.date) == m).all()

print(results)
```

```python
#  3. Convert the June temperatures to a list.
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temps_df = pd.DataFrame(results, columns=['date','june_temperature'])
june_temps_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
m = "12"
results = session.query(Measurement.date, Measurement.tobs).\
    filter(func.strftime("%m", Measurement.date) == m).all()

print(results)
```

```python
# 7. Convert the December temperatures to a list.
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temps_df = pd.DataFrame(results, columns=['date','dec_temperature'])
dec_temps_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temps_df.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')

fig, axes = plt.subplots(1,2, figsize=(10,6))

june_temps_df.hist('june_temperature', bins = 15, ax=axes[0], edgecolor='white')
dec_temps_df.hist('dec_temperature', bins = 15, ax=axes[1], color='green', edgecolor='white')
axes[0].set_xticks(np.arange(55,95, step=5))
axes[0].set_xlabel('deg F', fontsize = 16)
axes[1].set_xticks(np.arange(55,95, step=5))
axes[1].set_xlabel('deg F', fontsize=16)
```

```python
plt.style.use('seaborn-deep')
x = june_temps_df['june_temperature']
y = dec_temps_df['dec_temperature']

bins = np.linspace(56,86,25)

plt.subplots(figsize=(10,6))
plt.hist([x,y], bins, label=['june','dec'], color=['red','green'])
plt.legend(fontsize=16, loc='upper right')
plt.xlabel('deg F', fontsize=16)
plt.title("Oahu Temperatures in June and December", fontsize=20)
plt.ylabel('frequency', fontsize=16)
plt.show()
```
